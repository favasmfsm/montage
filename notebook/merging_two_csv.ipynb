```python
import pandas as pd
pd.set_option('display.max_columns', None)  # Display the full content of each column
```

```python
df1 = pd.read_csv("../data/DATA EXPORT FROM MATRIX - PACKAGE REPORT.csv")
```

```python
df = df[df.MatrixPackage==df.Package]
```
